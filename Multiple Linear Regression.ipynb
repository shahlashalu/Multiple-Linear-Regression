# **DUMMY COLUMN CREATION**

```python
df1 = pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']]).astype(int)
df1
```

```python
df1 = pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True).astype(int)
df1
```

```python
# CONCATINATION
df2 = pd.concat([df,df1],axis = 1)
df2
```

```python
df2.columns
```

```python
df = df2.drop(['Name','Unnamed: 0','Location','Fuel_Type','Transmission','Owner_Type'],axis = 1)
df
```

```python
df.head()
```

```python
import numpy as np
```

```python
df['Mileage'] = df['Mileage'].str.replace('km/kg','')
df['Mileage'] = df['Mileage'].str.replace('kmpl','')
df['Engine'] = df['Engine'].str.replace('CC','')
df['Power'] = df['Power'].str.replace('bhp','')
df['Power'] = df['Power'].str.replace('null','0')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Mileage'] = df['Mileage'].astype(float)
df['Engine'] = df['Engine'].astype(float)
df['Power'] = df['Power'].astype(float)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
#nan

import numpy as np
```

```python
df.loc[df.Power==0,'Power'] = np.NaN
```

```python
df['Engine'] = df['Engine'].fillna(df['Engine'].mean())
df['Power'] = df['Power'].fillna(df['Power'].mean())
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())
```

```python
df['Seats'].mode()
```

```python
df['Seats'] = df['Seats'].fillna(5.0)
```

```python
df.isnull().sum()
```

```python
x = df.drop(['Price'],axis = 1)
y = df['Price']
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print("r2_score",r2_score(y_test,y_pred))
```

```python
print("slope",regressor.coef_)
print("y intercept",regressor.intercept_)
```
